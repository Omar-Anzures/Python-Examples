```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Ejercicio:
    
    def __init__(self):
        self.x = [0,20,30]
        self.y = [20,40,0]
        self.x1 = [0,20,30]
        self.y1 = [40,0,30]
        
    def configuracion(self):
        plt.rc('figure',figsize = (11,10))
        opciones_fuentes ={'family': 'monospace','weight':'bold','size':'15'}
        plt.rc('font',**opciones_fuentes)
        plt.rc('lines', linewidth=5)
        
    def figura(self):
        self.figura = plt.figure()
        axis1 = self.figura.add_subplot(1,1,1) #hace que aparesca la figura
        axis1.set_xlabel('x - axis')
        axis1.set_ylabel('y - axis')
        axis1.set_title('Two or more lines with different widths and colors with suitable legends')
        return self.traza(axis1)
    
       
    def traza(self,axis1):
        axis1.plot(self.x,self.y,color='b',label='line1-width-3')
        axis1.plot(self.x1,self.y1,color='r',label='line2-width-5')
        axis1.legend(loc='upper right')
        return self.configuracion()
      
        
clase = Ejercicio()
clase.figura()

```
