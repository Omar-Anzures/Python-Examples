Pandas profiling:

Pandas profiling es un módulo open source de Python con el que podemos hacer rápidamente un análisis exploratorio de datos con solo unas pocas líneas de código. Además, por si esto no fuera suficiente para convencernos de la utilidad de la herramienta, esta genera informes interactivos en formato web que pueden ser presentados a cualquier persona, aunque no sepa programar.

```python
#Primer forma: Libreria Pandas profiling
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
data =  pd.read_csv("C:/Users/ovsantiago/Desktop/py/datasets/nba.csv")
```

```python
prof =  ProfileReport(data)
prof.to_file(output_file='C:/Users/ovsantiago/Desktop/py/reporte1.html')
```

```python
#En el modo mínimo se generará un informe simplificado con menos información que el completo, pero puede generarse con relativa rapidez para un gran conjunto de datos. 
profile_min = ProfileReport(data,minimal=True)
profile_min.to_file(output_file='C:/Users/ovsantiago/Desktop/py/reporte2.html')
```

Sweetviz:

Es una biblioteca de Python maravillosa y muy útil que nos proporciona la Analisis de datos exploratorios de un conjunto de datos dado con solo 2 líneas de código. Genera un informe de página HTML independiente con visualizaciones interactivas de un conjunto de datos. Puede ahorrarnos mucho tiempo que, de otro modo, habríamos dedicado a hacer EDA manualmente. También nos salva de errores que podríamos introducir al hacer las cosas por nosotros mismos.
Su funcion que mas me llamo la atencion es que puede comparar dos conjuntos de datos.

La función Compare () de Sweetviz se utiliza para comparar el conjunto de datos. Los comandos dados a continuación crearán y compararán nuestro conjunto de datos de prueba y entrenamiento.

```python
import sweetviz as sv
```

```python
result = sv.analyze(data)
```

```python
result.show_html('C:/Users/ovsantiago/Desktop/py/svreporte1.html')
```

```python
#Aqui se enpienza a comparar...
comp = sv.compare(data[229:],data[:229])
```

```python
comp.show_html('C:/Users/ovsantiago/Desktop/py/svreporte2.html')
```
