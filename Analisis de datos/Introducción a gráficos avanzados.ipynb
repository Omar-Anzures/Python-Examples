Matplotlib:

Es el paquete Python más utilizado en el ámbito de la ciencia de datos para representaciones gráficas.
Matplotlib es una librería de trazado utilizada para gráficos 2D en lenguaje de programación Python, es muy flexible y tiene muchos valores predeterminados incorporados que te ayudarán muchísimo en tú trabajo. Como tal, no necesitas mucho para comenzar, solamente tienes que hacer las importaciones necesarias, preparar algunos datos y con esto puedes comenzar a trazar tu función con la ayuda de la instrucción plot().

Hay diferentes formas de plotear con Matplotlib:

figure(num, figsize, dpi, facecolor, edgecolor, frameon)

    num = numeración de la figura, si num = None, las figuras se numeran automáticamente.

    figsize = w, h tuplas en pulgadas. Tamaño de la figura

    dpi = Resolución de la imagen en puntos por pulgada.

    facecolor = Color del rectángulo de la figura.

    edgecolor = Color del perímetro de la figura.

    frameon = Si es falso, elimina el marco de la figura.

Para crear mas figuras en una misma ventana podemos utilizar el siguiente comando:

subplot(numRows, numCols, plotNum)

    numRows = Número de filas
    numCols = Número de columnas
    plotNum = Número de gráfica

plot(x, y, linestyle, linewidth, marker) –> Permite incluir varias gráficas en una única figura.

    x = Abcisas.
    
    y = Ordenadas. Tanto x como y pueden ser abcisas tuplas, listas o arrays, pero ambas deben tener el mismo tamaño.
    
    linestyle = color y tipo de dibujar la gráfica. Por ejemplo ‘k- -‘
    
    linewidth = ancho de línea.
   
    marker = Marcador.



Formas y graficos
Las graficas pueden ser representadas en un objeto llamado figura, la cual puede tener una o mas representaciones.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#contenedor vacio
fig = plt.figure()
#se usa para agregar ejes a la figura como parte de un arreglo de subtrama.
#en pocas palabras ancho y alto del cuadrito
sub=fig.add_subplot(2,2,1)
#son los datos que va atener la grafica
sub.plot([3,6,-1,2])
#para que se visualiza
plt.show()
```

```python
#IMPRIMIR VARIOS GRAFICOS DIFERENTES

figura = plt.figure()

axis1=figura.add_subplot(2,2,1)
axis2=figura.add_subplot(2,2,2)
axis3=figura.add_subplot(2,2,3)

axis1.plot([3,6,-1,2])
axis2.hist(np.random.randn(100),bins=20)
axis3.scatter(np.arange(30), np.arange(30)+3*np.random.randn(30))
```

```python
#Hay otras maneras mas facil y recomendable para crear graficos el cual es:

figura2,axes = plt.subplots(2,3)
```

CAMBIOS PARA MEJORAR GRAFICOS(COLOR,TITULOS,ETC)

```python
#creacion de datos de prueba
data = np.random.randn(30).cumsum()
```

```python
#el metodo plt pasa tres parametos 
#1.parametro---datos que va a recibir
#2.parametro---el color que se quiere
#3.parametro---es el tipo de lineas que se muestra al unir los puntos
#4.parametro--son la forma que se quiere mostrar en los puntos
plt.plot(data,color='r',linestyle = 'dashed',marker='o')

"""En caso de que queramos cambiar la forma de las lineas como 
solidad o asi solo basta con cambiarle el nombre a linestyle:
°solid
°dashed
°dashdot
°dotted
"""
```

```python
plt.hist(data,color='g',linestyle='dashed')
```

```python
#se crea la figura (el cuadrito)
figura = plt.figure()

#se define el tamaño del cuadrito
ax = figura.add_subplot(1,1,1)

#podemos poner dos formas en la seccion de x numeros o texto
#en texto podemos poner como esta abajo con xticks y con xtickslabels texto
#ax.set_xticks([0,5,10,15,20])
ax.set_xticklabels(['cero','cinco','diez','quince','veinte','veinticinco','treinta'],rotation = 30,fontsize = 'medium')

#para poner el titulo a x y a y lo hacemos de la siguiente manera
ax.set_xlabel('Tiempo')
ax.set_ylabel('Variable')

#titulo general del diagrama
ax.set_title('Diagrama de prueba')


#se asigna los datos para que pueda representarlos en el cuadrito
#y se define el color con el tipo de lineas y los puntos.Label podemos poner un nombre del dato que se esta mostrando
ax.plot(data,color='black',linestyle = 'dashed',marker= 'o',label='random')

#agregamos la leyenda
ax.legend(loc='best')
```

Configurar estilos de fuente de la letra y el tamaño del la figura(el cuadrito xd):

```python
#se agrega el tamaño de la figura
plt.rc('figure',figsize = (7,5))
#tamaño y fuente de la letra
opciones_fuentes ={'family': 'monospace','weight':'bold','size':'15'}
#se sobrescribe el tamaño de la fuente de plt
plt.rc('font',**opciones_fuentes)
```

```python
#EJEMPLO

figura1 = plt.figure()
ax = figura1.add_subplot(1,1,1)
ax.set_xticklabels(['cero','cinco','diez','quince','veinte','veinticinco','treinta'],rotation = 30,fontsize = 'medium')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Variable')
ax.set_title('Diagrama de prueba')
ax.plot(data,color='black',linestyle = 'dashed',marker= 'o',label='random')
ax.legend(loc='best')
```
